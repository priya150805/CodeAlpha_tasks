```python
# TASK 3: Data Visualization

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Transform raw data into visual form
data = [
    {"Age": 22, "Salary": 25000, "Experience": 1, "Department": "IT"},
    {"Age": 25, "Salary": 30000, "Experience": 2, "Department": "HR"},
    {"Age": 28, "Salary": 40000, "Experience": 4, "Department": "Finance"},
    {"Age": 35, "Salary": 60000, "Experience": 8, "Department": "IT"},
    {"Age": 40, "Salary": 75000, "Experience": 10, "Department": "Finance"},
    {"Age": 30, "Salary": 45000, "Experience": 5, "Department": "HR"}
]

df = pd.DataFrame(data)

# Step 2: Use visualization tools
dept_salary = df.groupby("Department")["Salary"].mean()

# Step 3: Design clear visuals
dept_salary.plot(kind="bar")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()

# Step 4: Reveal insights
df["Experience"].hist()
plt.title("Experience Distribution")
plt.xlabel("Experience")
plt.ylabel("Count")
plt.show()

# Step 5: Support decision making
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
