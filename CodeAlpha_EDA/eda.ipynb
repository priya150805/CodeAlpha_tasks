```python
# TASK 2: Exploratory Data Analysis (EDA)


import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Ask meaningful questions
print("EDA Questions:")
print("1. Is salary related to experience?")
print("2. Are there missing values?")
print("3. What is the age distribution?")

# Step 2: Explore data structure
data = [
    {"Age": 22, "Salary": 25000, "Experience": 1, "Department": "IT"},
    {"Age": 25, "Salary": 30000, "Experience": 2, "Department": "HR"},
    {"Age": 28, "Salary": 40000, "Experience": 4, "Department": "Finance"},
    {"Age": 35, "Salary": 60000, "Experience": 8, "Department": "IT"},
    {"Age": 40, "Salary": 75000, "Experience": 10, "Department": "Finance"},
    {"Age": 30, "Salary": 45000, "Experience": 5, "Department": "HR"}
]

df = pd.DataFrame(data)

print("\nDataset Info:")
print(df.info())

# Step 3: Identify trends and patterns
print("\nStatistical Summary:")
print(df.describe())

# Step 4: Validate assumptions using visualization
plt.scatter(df["Experience"], df["Salary"])
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()

# Step 5: Detect data issues
print("\nMissing Values:")
print(df.isnull().sum())
```
